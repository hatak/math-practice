```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1.5
x = np.linspace(-1, 1)
y = a * x

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```python
def my_func(x):
    a = 3
    return x**a

x = np.linspace(0, 2)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```python
def my_func(x):
    return np.sqrt(x)

x = np.linspace(0, 9)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```python
def my_func(x):
    return 3*x**2 + 2*x + 1

x = np.linspace(-2, 2)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```python
def my_func(x):
    return 4*x**3 + 2*x**2 + x + 3

x = np.linspace(-2, 2)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```python
def my_sin(x):
    return np.sin(x)

def my_cos(y):
    return np.cos(y)

x = np.linspace(-np.pi, np.pi)
y_sin = my_sin(x)
y_cos = my_cos(x)

plt.plot(x, y_sin, label="sin")
plt.plot(x, y_cos, label="cos")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```

```python
def my_tan(x):
    return np.tan(x)

x = np.linspace(-1.3, 1.3)
y_tan = my_tan(x)

plt.plot(x, y_tan, label="tan")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```
